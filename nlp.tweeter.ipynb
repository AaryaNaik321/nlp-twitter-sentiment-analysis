```python
df = df.drop(['keyword','location','id'],axis=1)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
max(df['text'])
```

```python
from nltk.tokenize import word_tokenize
```

```python
df['text'].astype(object)
```

```python
import re
corpus =[]
for i  in range(len(df['text'])):
   # given condition that from a to z and A to Z keep all the charaters as it is other than it replace with space
    review= re.sub('[^a-zA-Z]',' ',df['text'][i])
    review= review.lower()
    corpus.append(review)
```

```python
corpus
```

```python
stopwords.words('english')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for i in corpus:
    #implsying tokenizer to the sentence
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemmatizer.lemmatize(word))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# create an instance of the CountVectorizer class
vectorizer = CountVectorizer()

# fit the vectorizer to your text data
X = vectorizer.fit_transform(corpus)
```

```python
y = df['target']
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
new_text = ["This is some new text data"]
new_text_transformed = vectorizer.transform(new_text)
y_pred = clf.predict(new_text_transformed)
```
